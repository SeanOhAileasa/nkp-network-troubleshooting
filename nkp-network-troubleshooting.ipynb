###### 2. Identifying the Problem
***

After completing this topic, you should be able to: ``identify a user's problem by following the network troubleshooting methodology``

All right, we'll begin this course by taking a look at implementing a network troubleshooting methodology. And troubleshooting can be difficult at times. So it's important to follow a logical troubleshooting methodology, whereby you essentially follow the same approach each time as best you can. So some common questions include, was the device functioning? In other words, did it work at any time or did it simply never work from the point that it was first taken out of the box, if you will? If it was functioning, then when was it last functioning? And has anything changed since the point that it was functioning to the point where it is no longer functioning? So clearly, you need to try to identify what the problem is. And the first goal of all troubleshooting is quite simply, what is the problem? And again, in some cases, that in and of itself can be challenging. But a user, for example, may have complained directly or communicated an issue to your help desk environment. But all networks in computing environments will require some sort of troubleshooting at some point in time. So again, gathering as much information as possible can help in terms of identifying the problem. So when it comes to gathering that information, try to implement a process with respect to anything to do with the incident. Now, this might require the use of any number of programs to make a proper assessment. For example, some simple tools such as ipconfig. If the user cannot connect to the network, you can verify what their IP address is. And of course, assess the state of the device upon arriving. Is it on, is it functioning, is it damaged? And, in essence, try to ascertain as much information as possible about what the state of that system was versus what the state is now. You can also try to duplicate the problem, attempt to recreate it on scene. And document any steps taken during this process. In other words, did the steps that you implemented fix all the problems? You may be able to identify additional issues during this phase of the process. Which, of course, you can use later on to try to determine if you can implement this same solution elsewhere should that problem arise again. Certainly question the users. Talk to them directly, but typically only after you complete your initial assessment of the problem. And this is due to the fact that, in some cases, users of course just don't have that much technical knowledge. So they might be trying to explain something to you. But they just might not know the correct term or even how to explain what the problem is. So typically, assess things on your own, then talk to them and try to flesh things out a little bit. They can absolutely be a good source of information. At least with respect to when things happen, what they were doing at the particular time. It can help to determine all of these issues with respect to when the problem was first discovered. So again, what was the behavior that they noticed? What programs were they using? Was that program ever used before, or it is a brand new program? Anything along those lines. So then when it comes to identifying the symptoms, of course, you need to try to identify what was actually happening. Try to determine the effects the problem is having on that specific device or on the network. And this information, of course, can be used later to determine if the actual problems have been resolved. In other words, did the symptoms actually stop? Also try to determine the changes. What exactly did change, and is that change good or bad? And, of course, determine if any new hardware or software has been installed that could be causing the problem. And, of course, this is with respect to since the last time everything worked as it should. And finally, approach problems individually. Isolate and address each problem one at a time. Try not to focus on the biggest picture right from the get go. Try to, in fact, do the opposite and just be specific with this particular problem, and then sort of grow out as necessary. By focusing on one problem at a time, you can ensure that any changes that you do implement are working. But if, of course, they don't work, then that's the point where you do need to start considering things beyond the scope of just this one particular problem. But typically, try not to start too far out, too far in terms of scope, and then try to work your way back down. It generally should be the other way around. So hopefully, you can come up with a process that works for you. Not every method of troubleshooting is suitable to every single person. Everybody has their preferences and the tools that they like. And even the processes that they find work best for them. But generally, try to incorporate all of those aspects into a troubleshooting methodology, and you should be able to get through most problems.

###### 3. Establishing the Cause
***

After completing this topic, you should be able to: ``establish a theory of probable cause by questioning the obvious and considering multiple approaches``

When it comes to effectively solving a problem, one of the first things you need to do is to identify the probable cause. You might be able to identify symptoms and even to address them, but typically, that results in a temporary fix only. If you don't get to the root cause, the symptoms are likely to reappear. So in this presentation, we'll take a look at establishing a theory of probable cause. Now, a network problem can be the result of many different causes. So, ideally, attempt to list all potential causes, and then prioritize that list to determine the likelihood of each one. This gives you the chance to eliminate possible causes while focusing in or identifying the probable cause, which might present the easiest solution. Now some of the questions that you might want to ask, of course, really should begin with the obvious. Is there power? Did the user do something, either unintentionally or intentionally? And can a simple reboot correct the issue? But in a more complex situation, the easiest solution is often the right one. But it might take a few tries to determine what that is, you know, what is the correct solution? It's not always obvious. So you really should consider multiple approaches, and this actually refers to the OSI Model. Try using a top-to-bottom or a bottom-to-top model approach. [An illustration of the OSI Model has seven layers, labeled layer one to layer seven, respectively. Each layer contains a corresponding component. Starting at layer one, the seven components in ascending order are as follows: Physical, Data Link, Network, Transport, Session, Presentation, and Application. ] For example, if a system is able to network but cannot get to the Internet, then you know it's not a physical problem. If I am seeing my file and print servers and my database connection and my internal mail server, then clearly, I have a physical network connection. But if I can't get to the Internet, well, maybe it's a problem with the default gateway, the router. Well, that's up in Layer 3. But if I then discover that in fact I can get to an Internet site by explicitly using an IP address as opposed to a name. And that's all the way up in Layer 7, DNS, an application protocol that is not functioning correctly. So if I cannot reach my DNS server, it is not going to be able to resolve a name for me. But if you use an IP address, you don't need to resolve the name. So that's an example of going from the bottom up. You might completely reverse that and go top down, that's fine. But ideally, when problem-solving, try these multiple directions. And you don't necessarily have to go from one extreme to the other. You might have a fairly good idea that it's maybe in the middle of the stack. But, simply put, try different approaches. Categorize the possible causes if you can. And these can be fairly general in terms of something that is not likely, likely, or most likely. So, again, in my example earlier, where a system was able to network, well, then physical is not likely at all. I can network, just certain things are not available to me. Again, the Internet makes the default gateway a likely cause. But then once we moved up and discovered that we could access a website by its IP address, then the DNS server becomes the most likely problem. And if you can, work as a team by dividing and conquering. Try assigning this particular possible cause to that team. They can investigate that much more deeply and then coordinate with other teams so that everyone is focusing on a specific task. And you can split up the troubleshooting tasks between your coworkers and test and implement a number of possible solutions much more quickly. And hopefully, you can get to the root of the problem as quickly as possible, so that things can return to normal and everyone can get back to work.

###### 4. Testing Techniques
***

After completing this topic, you should be able to: ``test a theory to determine the cause and next steps, and if escalation is required``

When it comes to troubleshooting, of course, the answer is not always obvious. So, in many cases, you simply need to come up with a theory, and of course, test that theory. So once you have questioned the obvious, a theory needs to be established if the obvious questions have not revealed what the problem is. So once the theory has been established, of course that then needs to be validated. So looking at a very simple example. If a user is having network connectivity issues, likely due to a faulty network cable, then of course the theory is, if we replace the cable, then of course the issue should be fixed. Now that in of itself needs to be verified, so you can't just put a new cable in and assume that everything is going to be fine. In other words, if a theory is proven to be incorrect, then clearly additional steps must be taken. So even given that earlier example, if you replaced the cable and there is still no connectivity, then you obviously have to look for a different cause. So a new theory of probable cause must be established. And perhaps you escalate, in some cases, if further testing still does not eliminate the issue. But again, going back to the original example, if the new cable didn't help, well, maybe the switch is down. So, if that's the case, then there should be multiple people who are experiencing connectivity issues. So clearly, that's something to find out as well, but ultimately it's all about just trying to figure out what you think is the most probable cause. And of course, test that out, then validate and/or verify if the problem has been solved. And if not, you move to the next level. And as mentioned, possibly escalate and get more people involved. But typically, by taking that step-by-step approach, you should be able to narrow down the focus of what could be the problem and hopefully alleviate that problem a little more quickly.

###### 5. Identifying Potential Effects
***

After completing this topic, you should be able to: ``establish a plan of action to resolve problems and identify potential effects``

Another consideration in the process of troubleshooting is identifying the potential effects of any given solution. In other words, once you have identified the problem, of course, you need to implement that solution. But a plan of action has to be formulated to identify possible or potential side effects, if you will, of that solution. And again, just going with a fairly simple example. If you do realize that a connectivity problem is related to something like a switch going down, then, of course, that switch should be replaced. But you might not be able to do that immediately. Maybe you have to order an entirely new switch, so it could be a while. So is there some kind of workaround that you could implement in the short-term, such as plugging them into a different switch? But that, of course, could have potential side effects. Because perhaps, you end up moving them to a different VLAN, or moving them off of the VLAN that they are supposed to be on. So clearly, you can't just assume to move them to a different switch without considering what that might result in, in terms of those effects. So planning becomes an important part of the troubleshooting process. You, in essence, have to consider, can I implement this solution right away? Or do we have to look at workarounds? And if there are workarounds, then what are the effects? So depending on the issues, it might require a more complex solution, of course. So consider writing out a step-by-step plan to ensure the best course of action is taken. And that plan should include the possible repercussions. Such as being disconnected for a certain amount of time versus being still connected but possibly in the wrong place. So, again, you need to try to identify what might happen with this particular implementation. So once the plan has been established, the solution can be implemented, of course, but ideally, this should resolve the issue right away. It might not, however, so you know, again, you'll also have to go back to the plan to see what might happen if we do this and/or what other options might be available. So if the solution that you implement hasn't fixed the problem, then of course, you might have to retrace your steps or attempt alternatives solutions that were outlined in the plan. So, again, maybe the plan said, well, let's move these people off to a different switch. But we soon discover that still results in a certain type of disconnectivity because now they can't reach a particular server, because of course, they were moved to a different VLAN. So go back to the plan, see what else you can do in terms of maybe reconnecting those people somewhere else. So, again, it depends, of course, on the complexity of the problem. In some cases, it is a very easy fix. For example, going all the way back to the scenario where it was simply a disconnected cable, it quite literally is, plug the cable back in and everyone is fine. So, you know, obviously, something like that does not really need a plan. But when there are larger, much broader problems, then you certainly need to consider what the potential effects might be of implementing your solution, and of course, considering alternative options.

###### 6. Resolving Problems
***

After completing this topic, you should be able to: ``implement an appropriate solution or escalate if required``

So continuing on with the troubleshooting process, the next step is, of course, to fully resolve the problem. But once you have identified the potential effects, you can and should implement the plan. But, of course, you also need to verify that the necessary authority has been granted to whomever is implementing the solution. In many cases, certain issues might fall outside of the scope of your own skillset. You may have identified what the problem is, you may have identified what the solution is and what the effects are. But you still might not actually have the ability or the permission to implement that plan. So if any given technician does not have authority or the skillset to resolve the problem, then, of course, it should be escalated. Now when escalating a problem, be sure to include all facts and documentation from the investigation. And this will help to ensure that the new technician, who is taking over the problem, does not have to redo any work that has already been completed. They don't want to have to start from scratch, in essence. So most organizations, typically, won't expect a single technician to know everything. They certainly will allow you to generally perform the tasks that are within your skillset, or at least within the authority that you've been granted. But in many cases, they do have this tiered structure, whereby they just simply don't want you to implement a solution where you might not be entirely familiar with what has to be done. So escalation procedures will certainly vary from organization to organization. And the general rule is that it should begin with the closest help. In other words, talk to the colleagues and notify whomever is your immediate supervisor. But, generally, don't go all the way to the top, if you will, for any given problem, just whoever is your immediate supervisor should be notified. Then, of course, they can handle it from that point. If it's deemed to be beyond their skillset or beyond their authority, then they in turn will notify their immediate supervisor, and it just builds from that point. So that, ultimately, you can resolve whatever the problem is with the appropriate person at some point. But as mentioned, in many cases, if you are fairly new to an environment, for example, they just will not give you the authority to correct every problem. Nor would they expect you to know the solution to every problem. Everyone gains experience as they go. So the longer you are involved in this process, undoubtedly, the higher up you will go, if you will. And you'll gain more permissions and more experience as you go, and, of course, be able to solve a broader range of problems.

###### 7. Verifying System Functionality
***

After completing this topic, you should be able to: ``verify full system functionality and implement preventative measures``

Continuing along with implementing a troubleshooting methodology, the next step is to verify system functionality. So once the plan has been implemented, the technician needs to verify full system functionality. In other words, did the solution correct the problem? Now, not only should the technician confirm that the original issue was resolved, but also that no new issues exist. This certainly won't happen all the time, but there is a possibility that fixing problem A can introduce problem B. So that's always something to be on the lookout for. So then upon validating the solution and confirming no new problems were introduced, you should also consider implementing some preventative measures to help ensure future problems are avoided. Now this, of course, depends on what the problem was. Certain things just cannot be avoided. Many hardware failures, for example, will happen. And in most cases you can only respond and/or react to those problems, you can't really prevent them. But if, for example, you discovered a virus on a system, and there was no anti-virus application, then clearly you should install an anti-virus program. And implement a schedule to ensure that virus definitions are updated frequently and that scans are happening frequently and perhaps even some user education, as well. Maybe they ran an executable from an untrusted email, so they should be informed that they should not do that. But again, it comes down to what the problem was. And if you can avoid it then certainly try your best to implement those preventative measures to reduce the amount of troubleshooting in the future.

###### 8. Documentation Processes
***

After completing this topic, you should be able to: ``document findings, actions, and outcomes``

Our final step of implementing a troubleshooting methodology is the documentation process. So after verifying that the solution is working effectively, it's important to update any troubleshooting documentation. Or, of course, begin new documentation if this particular problem has never been encountered before. But proper documentation is crucial as this information can be used in future instances or by other technicians. Which, of course, can save time if the same problem reoccurs. You are essentially creating a knowledge base so that instead of having to go through the entire troubleshooting methodology the next time this problem occurs, you simply consult what was done before. And you can, in essence, go straight to the correction. That does not guarantee that it will always work for every instance, but most of the time you can implement that solution and hopefully correct that problem. So the documentation should record the initial problem itself, the date and time, who reported it and, of course, the scope of how widespread the problem was. Of course, the symptoms, then the corrective actions that were taken, and, of course, the outcomes. And that should also include corrective actions that perhaps didn't work. If you tried this and it didn't work, then someone else can look at that and say, okay, well, chances are if I try the same thing, it probably won't work. So don't limit the documentation to only that which did work. List what you tried, list what failed, list what worked so that they can try other solutions if whatever they do try does not work. In essence, the more thorough the documentation, the easier it will be to resolve that problem in the future.
