###### 1. Identifying the Problem
***

After completing this topic, you should be able to: ``identify a user's problem by following the network troubleshooting methodology``.

< [Table of Contents](#top) | [References](#references) >
<a id="TroubleshootingMethodology"></a>
***
###### Troubleshooting Methodology
***

Implementing a network troubleshooting methodology. Troubleshooting can be difficult at times, so it's important to follow a logical troubleshooting methodology, whereby you essentially follow the same approach each time as best you can. 

< [Table of Contents](#top) | [References](#references) >
<a id="IdentifytheProblem"></a>
***
###### Identify the Problem
***

> Did it work at any time or did it simply never work from the point that it was first taken out of the box, if you will? 

If it was functioning, then:

> When was it last functioning? 

> Has anything changed since the point that it was functioning to the point where it is no longer functioning? 

Need to try to identify what the problem is and the first goal of all troubleshooting is quite simply, what is the problem? 

In some cases, that in and of itself can be challenging, but a user, for example, may have complained directly or communicated an issue to your help desk environment. 

All networks in computing environments will require some sort of troubleshooting at some point in time, so again, gathering as much information as possible can help in terms of identifying the problem. 

< [Table of Contents](#top) | [References](#references) >
<a id="GatherInformation"></a>
***
###### Gathering Information
***

When it comes to gathering that information, try to implement a process with respect to anything to do with the incident.

This might require the use of any number of programs to make a proper assessment, for example, some simple tools such as ``ipconfig``.

If the user cannot connect to the network, you can verify what their IP address is.

Assess the state of the device upon arriving. 

> Is it on?

> Is it functioning?

> Is it damaged?

In essence, try to ascertain as much information as possible about what the state of that system was versus what the state is now. 

< [Table of Contents](#top) | [References](#references) >
<a id="RecreatetheProblem"></a>
***
###### Recreate the Problem
***

Can also try to duplicate the problem, attempt to recreate it and document any steps taken during this process.

> Did the steps that you implemented fix all the problems?  

May be able to identify additional issues during this phase of the process, which, of course, you can use later on to try to determine if you can implement this same solution elsewhere should that problem arise again.

< [Table of Contents](#top) | [References](#references) >
<a id="QuestiontheUsers"></a>
***
###### Question the Users
***

Certainly question the users - talk to them directly, but typically only after the initial assessment of the problem is completed, this is due to the fact that, in some cases, users may not have that much technical knowledge, so they might be trying to explain something but just might not know the correct term or even how to explain what the problem is.

Typically, assess things on your own, then talk to them and try to flesh things out a little bit. 

They can absolutely be a good source of information, at least with respect to when things happen, what they were doing at the particular time - it can help to determine all of these issues with respect to when the problem was first discovered.

> What was the behavior that they noticed? 

> What programs were they using? 

> Was that program ever used before, or it is a brand new program? 

< [Table of Contents](#top) | [References](#references) >
<a id="IdentifytheSymptoms"></a>
***
###### Identify the Symptoms
***

When it comes to identifying the symptoms, of course, you need to try to identify what was actually happening. 

Try to determine the effects the problem is having on that specific device or on the network and this information, of course, can be used later to determine if the actual problems have been resolved. 

> Did the symptoms actually stop? 

< [Table of Contents](#top) | [References](#references) >
<a id="DeterminetheChanges"></a>
***
###### Determine the Changes
***

Try to determine the changes.

> What exactly did change?

> Is that change good or bad? 

Determine if any new hardware or software has been installed that could be causing the problem and, of course, this is with respect to since the last time everything worked as it should.

< [Table of Contents](#top) | [References](#references) >
<a id="ApproachProblemsIndividually"></a>
***
###### Approach Problems Individually
***

Approach problems individually - isolate and address each problem one at a time.

Try not to focus on the biggest picture right from the get go - do the opposite and be specific with the particular problem, and then sort of grow out as necessary.

By focusing on one problem at a time, you can ensure that any changes that you do implement are working. 

But if, of course, they don't work, then that's the point where you do need to start considering things beyond the scope of just this one particular problem. 

Typically, try not to start too far out, too far in terms of scope, and then try to work your way back down. 

It generally should be the other way around.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Summary
***

Not every method of troubleshooting is suitable to every single person - everybody has their preferences and the tools that they like. 

Even the processes that they find work best for them, but generally, try to incorporate all of those aspects into a troubleshooting methodology, and you should be able to get through most problems.

<a id="networkTroubleshooting3EstablishingtheCause"></a>

###### 2. Establish a Theory - Probable Cause
***

After completing this topic, you should be able to: ``establish a theory of probable cause by questioning the obvious and considering multiple approaches``.

When it comes to effectively solving a problem, one of the first things you need to do is to identify the probable cause. 

You might be able to identify symptoms and even to address them, but typically, that results in a temporary fix only, if you don't get to the root cause, the symptoms are likely to reappear. 

A network problem can be the result of many different causes, so, ideally, attempt to list all potential causes, and then prioritize that list to determine the likelihood of each one. 

This gives you the chance to eliminate possible causes while focusing in or identifying the probable cause, which might present the easiest solution. 

< [Table of Contents](#top) | [References](#references) >
<a id="QuestiontheObvious"></a>
***
###### Question the Obvious
***

Some of the questions that you might want to ask, of course, really should begin with the obvious. 

> Is there power? 

> Did the user do something, either unintentionally or intentionally? 

> Can a simple reboot correct the issue? 

But in a more complex situation, the easiest solution is often the right one, but it might take a few tries to determine what that is, you know, what is the correct solution - not always obvious. 

< [Table of Contents](#top) | [References](#references) >
<a id="ConsiderMultipleApproaches"></a>
***
###### Consider Multiple Approaches
***

Should consider multiple approaches, and this actually refers to the OSI Model - try using a top-to-bottom or a bottom-to-top model approach - each layer contains a corresponding component. 

If a system is able to network but cannot get to the Internet, then you know it's not a physical problem. 

If able to see my file and print servers and my database connection and my internal mail server, then clearly, I have a physical network connection. 

But if I can't get to the Internet, well, maybe it's a problem with the default gateway, the router - that's up in Layer 3. 

But if I then discover that in fact I can get to an Internet site by explicitly using an ``IP address`` as opposed to a ``name`` (that's all the way up in Layer 7 - ``DNS`` - application protocol).

Cannot reach my ``DNS`` server, then its not going to be able to resolve a ``name`` for me but if you use an ``IP address``, you don't need to resolve the name. 

So that's an example of going from the bottom up - might completely reverse that and go top down, that's fine. 

But ideally, when problem-solving, try these multiple directions and you don't necessarily have to go from one extreme to the other - might have a fairly good idea that it's maybe in the middle of the stack. 

Simply put, try different approaches. 

< [Table of Contents](#top) | [References](#references) >
<a id="CategorizePossibleCauses"></a>
***
###### Categorise Possible Causes
***

Categorize the possible causes if you can and these can be fairly general in terms of something that is not likely, likely, or most likely, so, again, in my example earlier, where a system was able to network, well, then physical is not likely at all - can network, just certain things are not available to me. 

Again, the Internet makes the default gateway a likely cause, but then once we moved up and discovered that we could access a website by its ``IP address``, then the ``DNS`` server becomes the most likely problem. 

< [Table of Contents](#top) | [References](#references) > 
<a id="DivideandConquerTeam"></a>
***
###### Divide and Conquer - Work as a Team
***

If possible, work as a team by dividing and conquering. 

Try assigning this particular possible cause to that team. 

They can investigate that much more deeply and then coordinate with other teams so that everyone is focusing on a specific task and you can split up the troubleshooting tasks between your coworkers and test and implement a number of possible solutions much more quickly and hopefully, you can get to the root of the problem as quickly as possible, so that things can return to normal and everyone can get back to work.

***
###### 3. Testing the Theory
***

After completing this topic, you should be able to: ``test a theory to determine the cause and next steps, and if escalation is required``.

When it comes to troubleshooting, of course, the answer is not always obvious, so, in many cases, you simply need to come up with a theory, and of course, test that theory. 

Once you have questioned the obvious, a theory needs to be established if the obvious questions have not revealed what the problem is - once the theory has been established, of course that then needs to be validated. 

If a user is having network connectivity issues, likely due to a faulty network cable, then of course the theory is, if we replace the cable, then of course the issue should be fixed - that in of itself needs to be verified, so you can't just put a new cable in and assume that everything is going to be fine. 

In other words, if a theory is proven to be incorrect, then clearly additional steps must be taken.

Even given that earlier example, if you replaced the cable and there is still no connectivity, then you obviously have to look for a different cause, so a new theory of probable cause must be established and perhaps you escalate, in some cases, if further testing still does not eliminate the issue - if the new cable didn't help, well, maybe the switch is down - if that's the case, then there should be multiple people experiencing connectivity issues, so clearly, that's something to find out as well.

Ultimately it's all about just trying to figure out what you think is the most probable cause and of course, test that out, then validate and/or verify if the problem has been solved and if not, you move to the next level and as mentioned, possibly escalate and get more people involved, but typically, by taking that step-by-step approach, you should be able to narrow down the focus of what could be the problem and hopefully alleviate that problem a little more quickly.

<a id="networkTroubleshooting5IdentifyingPotentialEffects"></a>
***
###### 4. Identifying Potential Effects
***

After completing this topic, you should be able to: ``establish a plan of action to resolve problems and identify potential effects``.

Another consideration in the process of troubleshooting is identifying the potential effects of any given solution, in other words, once you have identified the problem, of course, you need to implement that solution - a plan of action has to be formulated to identify possible or potential side effects, if you will, of that solution. 

If you do realize that a connectivity problem is related to something like a switch going down, then, of course, that switch should be replaced, but you might not be able to do that immediately. 

Maybe you have to order an entirely new switch, so it could be a while, so is there some kind of workaround that you could implement in the short-term.

> Plugging into a different switch? 

But that, of course, could have potential side effects because perhaps, you end up moving them to a different VLAN, or moving them off of the VLAN that they are supposed to be on, so clearly, you can't just assume to move them to a different switch without considering what that might result in, in terms of those effects, so planning becomes an important part of the troubleshooting process. 

> Can I implement this solution right away? 

> Do we have to look at workarounds? 

If there are workarounds:

> Then what are the effects? 

So depending on the issues, it might require a more complex solution, so consider writing out a step-by-step plan to ensure the best course of action is taken and that plan should include the possible repercussions, such as being disconnected for a certain amount of time versus being still connected but possibly in the wrong place - again, you need to try to identify what might happen with this particular implementation. 

Once the plan has been established, the solution can be implemented, of course, but ideally, this should resolve the issue right away - it might not, then have to go back to the plan to see what might happen if we do this and/or what other options might be available, so if the solution that you implement hasn't fixed the problem, then of course, you might have to retrace your steps or attempt alternatives solutions that were outlined in the plan - maybe the plan said, well, let's move these people off to a different switch - but we soon discover that still results in a certain type of disconnectivity because now they can't reach a particular server, because of course, they were moved to a different VLAN. 

So go back to the plan, see what else you can do in terms of maybe reconnecting those people somewhere else. 

Again it depends, of course, on the complexity of the problem, in some cases, it is a very easy fix, for example, going all the way back to the scenario where it was simply a disconnected cable, it quite literally requires plugging the cable back in and everyone is fine - obviously, something like that does not really need a plan. 

But when there are larger, much broader problems, then you certainly need to consider what the potential effects might be of implementing your solution, and of course, considering alternative options.

<a id="networkTroubleshooting6ResolvingProblems"></a>
***
###### 5. Resolving Problems
***

After completing this topic, you should be able to: ``implement an appropriate solution or escalate if required``.

The next step is, of course, to fully resolve the problem, but once you have identified the potential effects, you can and should implement the plan - of course, you also need to verify that the necessary authority has been granted to whomever is implementing the solution. 

In many cases, certain issues might fall outside of the scope of your own skillset - may have identified what the problem is, you may have identified what the solution is and what the effects are but you still might not actually have the ability or the permission to implement that plan. 

If any given technician does not have authority or the skillset to resolve the problem, then, of course, it should be escalated. 

Now when escalating a problem, be sure to include all facts and documentation from the investigation and this will help to ensure that the new technician, who is taking over the problem, does not have to redo any work that has already been completed - they don't want to have to start from scratch, in essence. 

So most organizations, typically, won't expect a single technician to know everything, they certainly will allow you to generally perform the tasks that are within your skillset, or at least within the authority that you've been granted but in many cases, they do have this tiered structure, whereby they just simply don't want you to implement a solution where you might not be entirely familiar with what has to be done. 

So escalation procedures will certainly vary from organization to organization and the general rule is that it should begin with the closest help, in other words, talk to the colleagues and notify whomever is your immediate supervisor but, generally, don't go all the way to the top, if you will, for any given problem, just whoever is your immediate supervisor should be notified then, of course, they can handle it from that point. 

If it's deemed to be beyond their skillset or beyond their authority, then they in turn will notify their immediate supervisor, and it just builds from that point, so that, ultimately, you can resolve whatever the problem is with the appropriate person at some point. 

As mentioned, in many cases, if you are fairly new to an environment, for example, they just will not give you the authority to correct every problem nor would they expect you to know the solution to every problem. 

Everyone gains experience as they go, so the longer you are involved in this process, undoubtedly, the higher up you will go, if you will and you'll gain more permissions and more experience as you go, and, of course, be able to solve a broader range of problems.

###### 6. Verifying System Functionality
***

After completing this topic, you should be able to: ``verify full system functionality and implement preventative measures``.

The next step is to verify system functionality, so once the plan has been implemented, the technician needs to verify full system functionality.

> Did the solution correct the problem? 

Not only should the technician confirm that the original issue was resolved, but also that no new issues exist - this certainly won't happen all the time, but there is a possibility that fixing problem A can introduce problem B, so that's always something to be on the lookout for. 

Then upon validating the solution and confirming no new problems were introduced, you should also consider implementing some preventative measures to help ensure future problems are avoided. 

Now this, of course, depends on what the problem was - certain things just cannot be avoided - many hardware failures, for example, will happen and in most cases you can only respond and/or react to those problems, you can't really prevent them but if, for example, you discovered a virus on a system, and there was no anti-virus application, then clearly you should install an anti-virus program and implement a schedule to ensure that virus definitions are updated frequently and that scans are happening frequently and perhaps even some user education, as well - maybe they ran an executable from an untrusted email, so they should be informed that they should not do that. 

Again, it comes down to what the problem was and if you can avoid it then certainly try your best to implement those preventative measures to reduce the amount of troubleshooting in the future.

***
###### 7. Documentation Processes
***

After completing this topic, you should be able to: ``document findings, actions, and outcomes``.

Our final step of implementing a troubleshooting methodology is the documentation process. 

After verifying that the solution is working effectively, it's important to update any troubleshooting documentation or, of course, begin new documentation if this particular problem has never been encountered before but proper documentation is crucial as this information can be used in future instances or by other technicians - can save time if the same problem reoccurs. 

You are essentially creating a knowledge base so that instead of having to go through the entire troubleshooting methodology the next time this problem occurs, you simply consult what was done before and you can, in essence, go straight to the correction. 

That does not guarantee that it will always work for every instance, but most of the time you can implement that solution and hopefully correct that problem.

So the documentation should record the initial problem itself, the date and time, who reported it and, of course, the scope of how widespread the problem was, of course, the symptoms, then the corrective actions that were taken, and, of course, the outcomes and that should also include corrective actions that perhaps didn't work - if you tried this and it didn't work, then someone else can look at that and say, well, chances are if I try the same thing, it probably won't work, so don't limit the documentation to only that which did work - list what you tried, list what failed, list what worked so that they can try other solutions if whatever they do try does not work. 

In essence, the more thorough the documentation, the easier it will be to resolve that problem in the future.

###### 1. Exercise: Apply Network Troubleshooting
***

i. Describe the steps used to identify a problem.

These can include gathering information, questioning the users, assessing the state of the system or the problem, attempting to duplicate the problem, identify what has changed, if anything, and, of course, identify the symptoms.

ii. List the steps used to establish the cause of a problem.

This can involve listing all possible causes. 

Then prioritizing that list, addressing the obvious.

> Is there power? 

Categorize the likelihood of each problem, considering multiple approaches, and possibly dividing into teams if it's a fairly widespread problem. 

iii. Describe the potential effects of a solution. 

This refers to the fact that, in some cases, you might implement a solution that, in itself, introduces new problems. 

Need to ensure that whatever solution you implement, in fact, does not cause any additional problems. 

iv. Describe some best practices of documentation. 

Includes recording the initial problem, the date and time, who reported the issue, the scope of the issue, the symptoms, of course, the actions that were taken, and the result of those actions. 

This helps you to build a knowledge base, so that if and/or when the same problem occurs in the future, you can simply refer to the knowledge base and avoid many of the same steps that you have to go through for each new problem.
